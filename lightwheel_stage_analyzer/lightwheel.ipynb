```python
def face_detector(pic, confidence_factor = 0.4):
  modelFile = "res10_300x300_ssd_iter_140000.caffemodel"
  configFile = "deploy.prototxt.txt"
  net = cv2.dnn.readNetFromCaffe(configFile, modelFile)
  img = cv2.imread(pic)
  h, w = img.shape[:2]
  blob = cv2.dnn.blobFromImage(cv2.resize(img, (300, 300)), 1.0,
  (300, 300), (104.0, 117.0, 123.0))
  net.setInput(blob)
  faces = net.forward()
  #to draw faces on image
  for i in range(faces.shape[2]):
          confidence = faces[0, 0, i, 2]
          if confidence > confidence_factor:
              box = faces[0, 0, i, 3:7] * np.array([w, h, w, h])
              (x, y, x1, y1) = box.astype("int")
              cv2.rectangle(img, (x, y), (x1, y1), (0, 0, 255), 2)
              # cv2.imshow(img)
              return img
          else:
              return 0
```

```python
import cv2
import numpy as np


modelFile = "res10_300x300_ssd_iter_140000.caffemodel"
configFile = "deploy.prototxt.txt"
net = cv2.dnn.readNetFromCaffe(configFile, modelFile)

confidence_factor = 0.6   
```

```python
vidcap = cv2.VideoCapture('C:/Users/BMH/Desktop/TSS/test data/124_CP_mahla/start_wheel_cam2.avi')
success,image = vidcap.read()

if vidcap.isOpened():
  width  = vidcap.get(cv2.CAP_PROP_FRAME_WIDTH)
  height = vidcap.get(cv2.CAP_PROP_FRAME_HEIGHT)
  fps = vidcap.get(cv2.CAP_PROP_FPS)
  print(width,height,fps)

out = cv2.VideoWriter('./124_CP_mahla.avi', 0, fps, (int(width),int(height)))
```

```python

frame_num = 0
face_detected_frames = 0

while success:
  
    success,img = vidcap.read()

    if(not success):
      break

    h, w = img.shape[:2]
    blob = cv2.dnn.blobFromImage(cv2.resize(img, (300, 300)), 1.0,
    (300, 300), (104.0, 117.0, 123.0))
    net.setInput(blob)
    faces = net.forward()
    #to draw faces on image
    for i in range(faces.shape[2]):
          confidence = faces[0, 0, i, 2]
          if confidence > confidence_factor:
              box = faces[0, 0, i, 3:7] * np.array([w, h, w, h])
              (x, y, x1, y1) = box.astype("int")
              cv2.rectangle(img, (x, y), (x1, y1), (0, 0, 255), 2)
              # cv2_imshow(img)
              face_detected_frames += 1
    out.write(img)
#     cv2.imwrite("/content/87/frame%d.jpg" % frame_num, img)

    frame_num += 1

    print("frame number: ",frame_num,"  ","face_detected_frames:  ",face_detected_frames)


vidcap.release()
out.release()
```

```python
print("total frames: %s, detected frames: %s, ratio: %s" %(frame_num, face_detected_frames, (face_detected_frames/frame_num)))
```
